```python
from datasets import get_dataset_config_names

xtreme_subsets = get_dataset_config_names("xtreme")
print(f"XTREME has {len(xtreme_subsets)} configurations")
```

```python
print(xtreme_subsets)
```

```python
panx_subsets = [s for s in xtreme_subsets if s.startswith("PAN")]
print(len(panx_subsets))
panx_subsets
```

```python
import pandas as pd
from datasets import load_dataset

count = dict()
ds = dict()

for lang in panx_subsets:
    ds[lang] = load_dataset("xtreme", name=lang)
    count[lang] = dict()
    count[lang]["train"] = len(ds[lang]["train"])
    count[lang]["test"] = len(ds[lang]["test"])
    count[lang]["validation"] = len(ds[lang]["validation"])

pd.DataFrame.from_dict(count, orient="index")
```

```python
tags = ds["PAN-X.tr"]["train"].features["ner_tags"].feature
print(tags)
```

```python
tokens = ds["PAN-X.tr"]["train"][0]["tokens"]
label_indices = ds["PAN-X.tr"]["train"][0]["ner_tags"]
labels = [tags.names[p] for p in label_indices]

df = pd.DataFrame(data=[tokens, labels], index=['Tokens', 'Tags'])
df
```
